## Objetivo:

Hacer un clustering de libros basado en la información de la descripción de cada libro.

```python
textos = list(df['book_desc'].values)
textos = [str(t) for t in textos]
textos

#tokenizar cada texto (descripcion de cada libro)
tokens = []
for t in textos:
    tokens.append(word_tokenize(t))

tokens
```

```python

#eliminar stopwords
stop_words = set(stopwords.words('english'))
stop_words.update(STOP_WORDS)
stop_words.update(punctuation)

tokens_clean = []
for t in tokens:
    tokens_clean.append([w for w in t if w not in stop_words])

tokens_clean
```

```python
#lematizar
tokens_lem = []
for t in tokens_clean:
    tokens_lem.append([nlp(w)[0].lemma_ for w in t])

tokens_lem
```
